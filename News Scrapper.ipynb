```python
import requests
API_KEY = "Yourapikey"
SEARCH_ENGINE_ID = "Yourid"

```

### The code below is used to search a term. Go ahead and add your search term in the query object.  üîç<br>
<li>The page variable is used to make a note of page you are in. Initially lets be on page 1.üìÑ</li>
<li> start is the number which will get you articles of the specific number, we want to increament it by 10</li>


```python
query = "Assam Flood 2020"
page = 1
start = (page - 1) * 10 + 1
url = f"https://www.googleapis.com/customsearch/v1?key={API_KEY}&cx={SEARCH_ENGINE_ID}&q={query}&start={start}"
data = requests.get(url).json()
```

# The main algorithm

```python
news=[]
search_items = data.get("items")
while page<11:
    search_items = data.get("items")
    for i, search_item in enumerate(search_items, start=1):
        title = search_item.get("title")
        snippet = search_item.get("snippet")
        html_snippet = search_item.get("htmlSnippet")
        link = search_item.get("link")
        Data= {
            'Title':title,
            'Snippet':snippet,
            'htmlSnippet':html_snippet,
            'link':link
        }
        news.append(Data)
        print("="*10, f"Result #{i+start-1}", "="*10)
        print("Title:", title)
        print("Description:", snippet)
        print("URL:", link, "\n")
    page=page+1
    start = (page - 1) * 10 + 1
    url = f"https://www.googleapis.com/customsearch/v1?key={API_KEY}&cx={SEARCH_ENGINE_ID}&q={query}&start={start}"
    data = requests.get(url).json()
```

### Now lets import pandas so that we can save the list into a dataframe

```python
import pandas as pd
```

```python
news=pd.DataFrame(news,columns=['Title','Snippet','htmlSnippet','link'])
```
